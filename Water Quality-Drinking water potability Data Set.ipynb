## Are there any data quality issues like typos, outliers, or inconsistent formatting?

```python
import pandas as pd

# Load dataset
df = pd.read_csv("water_potability_eda.csv")

# Select numeric columns
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Check outliers column-wise
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col}: {outliers.shape[0]} outliers")
```

## Fix Inconsistencies in Categorical Data

**Review text data for inconsistency in spelling, case (uppercase/lowercase), or formatting**



```python
import pandas as pd
df = pd.read_csv("water_potability_eda.csv")
cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    df[col] = (
        df[col]
        .astype(str)        # handle missing values
        .str.strip()        # remove leading/trailing spaces
        .str.lower()        # unify case (Yes, YES â†’ yes)
        .str.replace(r"\s+", " ", regex=True)  # fix extra spaces
    )
value_map = {
    "y": "yes",
    "yes": "yes",
    "n": "no",
    "no": "no"
}

for col in cat_cols:
    df[col] = df[col].replace(value_map)
for col in cat_cols:
    print(f"\n{col} unique values:")
    print(df[col].unique())
```

**Remove unnecessary whitespace or special characters**

```python
import pandas as pd

# Load dataset
df = pd.read_csv("water_potability_eda.csv")

# Identify categorical (text) columns
cat_cols = df.select_dtypes(include="object").columns

# Check if categorical columns exist
if len(cat_cols) == 0:
    print("No categorical (text) columns found to clean.")
else:
    for col in cat_cols:
        df[col] = (
            df[col]
            .fillna("")                         # handle missing values safely
            .str.strip()                       # remove leading/trailing spaces
            .str.replace(r"\s+", " ", regex=True)   # remove extra spaces
            .str.replace(r"[^a-zA-Z0-9 ]", "", regex=True)  # remove special characters
        )

    # Verify output
    for col in cat_cols:
        print(f"\n{col} unique values after cleaning:")
        print(df[col].unique())
```

## Observation:

There are no categorical (text) columns found to clean.

**Standardize categories to avoid fragmentation of groups.**

```python
import pandas as pd
df = pd.read_csv("water_potability_eda.csv")
cat_cols = df.select_dtypes(include="object").columns
if len(cat_cols) == 0:
    print("No categorical columns found to standardize.")
else:
    standard_maps = {
        "yes": ["yes", "y", "yes ", "YES", "Yes"],
        "no": ["no", "n", "no ", "NO", "No"],
        "male": ["m", "male", "Male", "MALE"],
        "female": ["f", "female", "Female", "FEMALE"]
    }
    for col in cat_cols:
        df[col] = (
            df[col]
            .fillna("")
            .str.strip()
            .str.lower()
        )
        for standard_value, variants in standard_maps.items():
            df[col] = df[col].replace(variants, standard_value)
    for col in cat_cols:
        print(f"\n{col} categories after standardization:")
        print(df[col].value_counts())
```

## Observation

Hence No categorical columns found to standardize.

**Convert date fields to datetime**

```python
import pandas as pd
df = pd.read_csv("water_potability_eda.csv")
print("Before conversion:\n")
print(df.dtypes)
```

```python
date_columns = ["date", "created_date", "updated_date"]

for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")
for col in df.columns:
    if df[col].dtype == "object":
        try:
            df[col] = pd.to_datetime(df[col], errors="raise")
        except:
            pass
```

```python
print("\nAfter conversion:\n")
print(df.dtypes)
```

**Mark repeated label/text columns as 'categorical' where appropriate.**

```python
import pandas as pd
df = pd.read_csv("water_potability_eda.csv")
cat_cols = df.select_dtypes(include="object").columns
if len(cat_cols) == 0:
    print("No text columns found to convert into categorical type.")
else:
    for col in cat_cols:
        unique_ratio = df[col].nunique() / len(df)
        if unique_ratio < 0.5:
            df[col] = df[col].astype("category")
            print(f"Column '{col}' converted to categorical.")
        else:
            print(f"Column '{col}' left as object (high uniqueness).")
print("\nUpdated data types:\n")
print(df.dtypes)
```
